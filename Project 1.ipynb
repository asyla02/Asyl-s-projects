```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Load Data

```python
phone=pd.read_excel('Phones.xlsx','Global')
phone.head()
```

# Data Inspection

```python
phone['2016'].dtype
```

```python
for y in range(2008,2017):
    phone[str(y)]=pd.to_numeric(phone[str(y)], errors='coerce')
phone.head()
```

```python
print(phone.shape)
phone=phone.loc[phone['Brand Name']!='Total']
print(phone.shape)
phone.head()
```

```python
phone.loc[phone['Geography']=='India']['Brand Name'].unique()
```

# Compute Market Shares

```python
world=phone
```

```python
world_country=world.groupby(['Geography'])[['2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017']].sum()
world_country.head()
```

```python
world_country_brand=world.groupby(['Geography','Brand Name'])[['2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017']].sum()
world_country_brand
```

```python
brand_country_share=100*world_country_brand.div(world_country, level='Geography')
brand_country_share
```

```python
brand_country_share.loc['China'].sort_values('2016',ascending=False).transpose()
```

# Charting

```python
china_share=brand_country_share.loc['China'].sort_values('2016',ascending=False).transpose()
china_share.head()
```

```python
china_share[['Huawei','iPhone','Samsung']].plot.bar()
plt.title('Mobile Phones Market in China')
plt.xlabel('Year')
plt.ylabel('Market Share in %')
```

```python
china_share[['Huawei','iPhone','Samsung']].plot.bar(stacked=True)
plt.title('Mobile Phones Market in China')
plt.xlabel('Year')
plt.ylabel('Market Share in %')
```

# Merging Datasets

```python
china=brand_country_share.loc['China']['2016'][['Samsung','Huawei','iPhone']].reset_index()
china
```

```python
india=brand_country_share.loc['India']['2016'][['Samsung','Huawei','iPhone']].reset_index()
india
```

```python
share2016=china.merge(india, left_on='Brand Name', right_on= 'Brand Name')
share2016.columns=['Brand','China','India']
share2016
```
