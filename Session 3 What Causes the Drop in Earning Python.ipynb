```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

# Load Excel Data

```python
data=pd.read_excel('Session 3 What Causes the Drop in Earning.xlsx', 'Data Collected by Firm 1')
print(data.shape)
data.head(10)
```

# Compute for Our Firm

```python
we=data.loc[(data['Firm ID']==1) & (data['Industry Sector']==4)]
we
```

```python
we['Cost of Good Sold/Sales Revenue']=we['Cost of Good Sold']/we['Sales Revenue']
we['Marketing Expense/Sales Revenue']=we['Marketing Expense']/we['Sales Revenue']
we['Other SG&A/Sales Revenue']=we['Other SG&A']/we['Sales Revenue']
we
```

```python
we=we[['Year', 
       'Sales Revenue', 
       'Cost of Good Sold/Sales Revenue', 
       'Marketing Expense/Sales Revenue', 
       'Other SG&A/Sales Revenue']]
we
```

```python
we=we.set_index('Year').transpose()
we
```

# Compute Industry Aggregation

```python
industry=data.loc[data['Industry Sector']==4]
industry
```

```python
industry['Cost of Good Sold/Sales Revenue']=industry['Cost of Good Sold']/industry['Sales Revenue']
industry['Marketing Expense/Sales Revenue']=industry['Marketing Expense']/industry['Sales Revenue']
industry['Other SG&A/Sales Revenue']=industry['Other SG&A']/industry['Sales Revenue']
industry.head()
```

```python
industry2=industry.groupby('Year')['Cost of Good Sold/Sales Revenue', 
       'Marketing Expense/Sales Revenue', 
       'Other SG&A/Sales Revenue'].mean()
industry2
```

```python
industry2['Sales Revenue']=industry.groupby('Year')['Sales Revenue'].mean()
industry=industry2.transpose()
industry
```
